### Task A: 5 points

```python
unique_data = data['itemDescription'].unique()
print("\nUnique Items available in the Dataset:")
unique_data
```

### Task B: 7 points

Create a list of lists with grouping on member number and the transaction date and store it in a dataframe

```python
data['Date'] = pd.to_datetime(data['Date'])

grouped_dataset = data.groupby(['Member_number', 'Date'])['itemDescription'].apply(list).reset_index(name='Items')
print("\nGrouped Data:")
grouped_dataset
```

### Task C: 7 points

Transform the data having unique product as column name and indication of the product in the each transaction as 0/1.

```python
data['Date'] = pd.to_datetime(data['Date'])

encoded_dataset = grouped_dataset.set_index(['Member_number', 'Date'])['Items'].str.join('|').str.get_dummies().reset_index()
print("\nEncoded Data:")
encoded_dataset
```

### Task D: 7 points

Apply apriori algorithm with 0.001 support

```python

from mlxtend.frequent_patterns import apriori

apriori_algo = apriori(encoded_dataset.drop(['Member_number', 'Date'], axis=1), min_support=0.001, use_colnames=True)
print("\nFrequent Itemsets (Apriori):")
apriori_algo
```

### Task E: 5 points

Print top 5 association rules for lift metric and 1 threshold in accordance with support value.

```python
association_rules_apriori_algo = association_rules(apriori_algo, metric="lift", min_threshold=1)
top_5_rules_apriori_algorithm = association_rules_apriori_algo.nlargest(5, 'lift')
print("\nTop 5 Association Rules (Apriori):")
top_5_rules_apriori_algorithm
```

### Task F: 3 points

Store the association rules from top to bottom with respect to lift parameter and store it in a variable.

```python
sorted_rules_apriori_algorithm = association_rules_apriori_algo.sort_values(by='lift', ascending=False)
sorted_rules_apriori_algorithm
```

### Task G: 10 points

Understand this,
- antecedent support: If X is called antecendent, 'antecedent support' computes the proportion of transactions that contain the antecedent X.
- consequent support: If Y is called consequent, 'consequent support' computes the proportion of transactions that contain the antecedent Y.
- support: 'support' computes the proportion of transactions that contain the antecedent X and Y.
- confidence: Probability of buying Y when X is bought.
- lift: Represents how many times the probability of getting Y increases when X is received.

Define a function which takes the (sorted values from the association rules,a product name) and returns top recommendation list for that product name.

Tip: the recommendated item are Consequents by matching the product antecedents as product to search for.

```python
def get_top_recommendations(dataset_rules, product_name):
    antecedents = dataset_rules[dataset_rules['antecedents'].apply(lambda x: product_name in x)]
    recommendations_dataset_rules = antecedents['consequents'].apply(lambda x: list(x)[0] if len(x) == 1 else None).dropna()
    return recommendations_dataset_rules
```

### Task H: 5 points

Print top 5 reccomendated items for product 'sausage'.

```python
top_recommendations_sausage = get_top_recommendations(sorted_rules_apriori_algorithm, 'sausage')
print("\nTop 5 Recommended Items for 'sausage' (Apriori):")
top_recommendations_sausage.head()
```

### Task I: 5 points

Print top 5 reccomendated items for product 'sausage' with 0.005 support.

```python
apriori_algo_with_threshold = apriori(encoded_dataset.drop(['Member_number', 'Date'], axis=1), min_support=0.005, use_colnames=True)
association_rules_apriori_algo_with_threshold = association_rules(apriori_algo_with_threshold, metric="lift", min_threshold=1)

sorted_rules_apriori_algorithm_with_threshold = association_rules_apriori_algo_with_threshold.sort_values(by='lift', ascending=False)
top_recommendations_sausage_with_threshold = get_top_recommendations(sorted_rules_apriori_algorithm_with_threshold, 'sausage')

print("\nTop 5 Recommended Items for 'sausage' with support threshold 0.005 (Apriori):")
print(top_recommendations_sausage_with_threshold.head())
```

### Task J: 10 points

Perform task D through I except task G using FP-Growth algorithm.

```python
from mlxtend.frequent_patterns import fpgrowth

fpgrowth_freq_items = fpgrowth(encoded_dataset.drop(['Member_number', 'Date'], axis=1), min_support=0.001, use_colnames=True)

print("\nFrequent Itemsets (FP-Growth):")
print(fpgrowth_freq_items)

association_rules_fpgrowth = association_rules(fpgrowth_freq_items, metric="lift", min_threshold=1)
fpgrowth_rules = association_rules_fpgrowth.sort_values(by='lift', ascending=False)
```

### Task K: 8 points

Print Top 3 recommendations for below products using Apriori with 0.0015 support
- citrus fruit
- whole milk
- soda

```python
def top_recommendations(dataset_rules, product_name, n=3):
    antecedents = dataset_rules[dataset_rules['antecedents'].apply(lambda x: product_name in x)]
    recommendations_dataset_rules = antecedents['consequents'].apply(lambda x: list(x)[0] if len(x) == 1 else None).dropna()
    return recommendations_dataset_rules.head(n)

recommend_apriori_products = ['citrus fruit', 'whole milk', 'soda']
print("\nTop 3 Recommendations using Apriori with support threshold 0.0015:")
for product in recommend_apriori_products:
    recommendations_dataset_rules = top_recommendations(sorted_rules_apriori_algorithm, product, n=3)
    print(f"Recommended items for '{product}':")
    print(recommendations_dataset_rules)
```

### Task L: 8 points

Print Top 3 recommendations for below products using FP-Growth algorithm with 0.0015 support
- canned beer
- yogurt
- pastry

```python
recommend_apriori_products = ['canned beer', 'yogurt', 'pastry']

print("\nTop 3 Recommendations using FP-Growth with support threshold 0.0015:")

for product in recommend_apriori_products:
    recommendations_dataset_rules = top_recommendations(fpgrowth_rules, product, n=3)
    print(f"Recommended items for '{product}':")
    print(recommendations_dataset_rules)
```
